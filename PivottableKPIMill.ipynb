```python
import pandas as pd

# Menggunakan read_excel untuk file Excel asli
data = pd.read_excel('/content/Kebun_Tiga_Juhar_KPI.xls')

data.head()
```

```python
# 1. Menyaring data: Hanya mengambil baris yang kolom 'penilaian_header'-nya berisi 'S2'
data_s2 = data[data['penilaian_header'] == 'S2']

# 2. Membuat pivot table dari data yang sudah disaring tadi
pivot_kpi_s2 = pd.pivot_table(
    data_s2,                     # Menggunakan data_s2 yang sudah disaring
    index='textkpi',             # Row (Baris)
    columns='namakaryawan',      # Kolom
    values='target_detail1',     # Sigma Values
    aggfunc='sum',               # Menggunakan fungsi SUM (Total)
    fill_value=0                 # Mengubah data kosong menjadi 0
).reset_index()

# 3. Menampilkan hasil pivot di layar Colab
pivot_kpi_s2
```

```python
pivot_kpi_s2.to_excel('Pivot_Target_Karyawan_Tigajuhar_S2.xlsx', index=False)
```

```python
# 1. Membuat pivot table dengan Sigma Values: realisasi
pivot_realisasi_s2 = pd.pivot_table(
    data_s2,                     # Tetap menggunakan data yang sudah disaring S2
    index='textkpi',             # Row (Baris)
    columns='namakaryawan',      # Kolom ke samping
    values='realisasi',          # Sigma Values diganti menjadi kolom realisasi
    aggfunc='sum',               # Menggunakan fungsi SUM (Total)
    fill_value=0                 # Mengubah data kosong menjadi 0
).reset_index()

# 2. Menampilkan hasil pivot realisasi S2 di layar
pivot_realisasi_s2
```

```python
pivot_realisasi_s2.to_excel('Pivot_Realisasi_Karyawan_Tigajuhar_S2.xlsx', index=False)
```
