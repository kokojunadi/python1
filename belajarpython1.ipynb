```python
# Membuat Pivot Table dan hasilnya disimpan ke dalam nama baru: 'hasil_pivot'
hasil_pivot = pd.pivot_table(
    data,
    index='namajabatan',
    values='nilaiakhir',
    aggfunc='mean'
).reset_index()

# Menampilkan hasil pivot ke layar
hasil_pivot
```

```python
# Membuat pivot table sesuai permintaan Anda
pivot_kpi_karyawan = pd.pivot_table(
    data,
    index='textkpi',           # Row (Baris)
    columns='namakaryawan',    # Kolom
    values='target_detail1',   # Sigma Values
    aggfunc='sum',             # Menggunakan fungsi SUM (Total) jika ada data yang sama
    fill_value=0               # Mengubah data kosong (NaN) menjadi angka 0 agar rapi seperti di Excel
).reset_index()

# Menampilkan hasil pivot
pivot_kpi_karyawan
```
